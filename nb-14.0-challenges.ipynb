### Challenge 1: 
Execute the code cell below to see an example of how it works. 
Use markdown in the cell after the code-block to describe the function `random_words_api`. Try to be descriptive about what each step of code in this function does, and why it works. 

```python
def random_words_api(nwords=10):
    "no docstring"
    URL = "https://random-word-api.herokuapp.com/word"
    response = requests.get(url=URL, params={"number": nwords})
    return response.json()

# demonstration
random_words_api(5)
```

The random_words_api is a function that can take in an integer as a parameter or uses 10 as a default. 
It stores a weblink https://random-word-api.herokuapp.com/word as a variable named URL.
Using the requests module, it fetches the data from the weblink with the specified number to the website and stores it in a variable named response. 
The function then returns the result in json format.

Calling the function with 5 specified, returns 5 random words.

### Challenge 2: 
Use the `random_words_api` function to get 50 random words and store the result as a variable. Write a function that takes the list of words as input and returns a dictionary with the longest word as the key and the length of the longest word as a value. If there is a tie in the length of words then have it return additional words as keys with their lengths as values.

```python
#storing 50 words into a variable
temp1 = random_words_api(50)

#function to take the list of words as input and stores longest word as value in a key-value. 
def longest_word(x):
    
    #sorts the list with longest word in first index
    x.sort()
    
    #if the length of the first word is not the same as the second, only have to return first word
    if len(x[0]) != len(x[1]):
        longdict = {x[0]: len(x[0])}
        return longdict
    
    #else, the function traverses the list ensuring all the words tied for longest length is added to the dictionary
    else:
        longdict = {x[0]: len(x[0])}
        for a in range(len(x)):
            if len(x[a]) == len(x[a+1]):
                longdict.update({x[a+1]: len(x[a+1])})
            else:
                return longdict
        return longdict
    
longest_word(temp1)
```

### Challenge 3: 
Write a function to take the list of words as input and trim all words to be at most 5 characters in length, and return as a list.

```python
def trim_list(x):
    y = []
    for a in range(len(x)):
        if len(x[a])<5:
            y.append(a)
        else:    
            s=x[a][:5]
            y.append(s)
    return y
trim_list(temp1)
```

### Challenge 4: 
Write a function to take a list of words as input and to count the occurrence of all letters in every word and return as a dictionary mapping letters to integers., e.g., {'a': 10, 'b': 3, 'c': 5, ...}.  

```python

def letter_count(x):
    counts = dict()
    for a in x:
        for letter in a:
            keys = counts.keys()
            if letter in keys:
                counts[letter]+=1
            else:
                counts[letter]=1

    return counts  

letter_count(temp1)
```

### Challenge 5:
Use [toyplot](https://toyplot.readthedocs.io/en/stable/tutorial.html) to create a barplot of the occurrences of each letter in your dictionary from the previous challenge. This will represent a histogram of the letters. Play with the size and color of the figure to try to make it look nice.

```python
import toyplot.locator
def letter_histogram(counts):
    x = list(counts.keys())
    y = list(counts.values())
    
    canvas = toyplot.Canvas(width=600, height=600)
    axes = canvas.cartesian()
    axes.x.ticks.show = True
    axes.x.ticks.locator = toyplot.locator.Explicit(
    None,
    x)
    
       
    mark = axes.bars(range(len(x)), y, color="steelblue", style = {"stroke":toyplot.color.black})
    return mark

letter_histogram(letter_count(temp1))
```
